# 02. Exploratory Data Analysis (EDA)
Analyzing trends, seasonality, and relationships.

```python
import sys
import matplotlib.pyplot as plt
import seaborn as sns
sys.path.append('../')
from src.data_loader import load_time_series, load_weather_data
from src.preprocessing import preprocess_pipeline

base_dir = '../'
df_ts = load_time_series(base_dir)
df_weather = load_weather_data(base_dir)
df = preprocess_pipeline(df_ts, df_weather)
```

```python
# Time Series Plot
plt.figure(figsize=(15, 6))
df['load_actual'].plot()
plt.title('Electrical Load over Time')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap')
plt.show()
```
